```python
def get_sec(tstr):
    if(isinstance(tstr, str)):
        v = tstr.split(":")
        m, s = v[0], v[1]
    else:
        m, s = tstr.hour, tstr.minute
    if(int(m) < 10):
        m = '0'+str(int(m))
    if(int(s) < 10):
        s = '0'+str(int(s))
    return "00:"+str(m)+":"+str(s)

import os
import pandas as pd
import subprocess


train_list = []
test_list = []
for l in open('data (1)\\train.csv'):
    label = int(l[-2])

    id = int(l[len('/home/swli2/slowfast/data/video'):-len('_final.mp4 x\n')])

    train_list.append(id)

for l in open('data (1)\\test.csv'):
    label = int(l[-2])
    if(l[-3:-1] == '10'):
        id = int(l[len('/home/swli2/slowfast/data/video'):-len('_final.mp4 xx\n')])
    else:
        id = int(l[len('/home/swli2/slowfast/data/video'):-len('_final.mp4 x\n')])

    test_list.append(id)



df = pd.read_excel('PD Data doc.xlsx')
for i in range(df.shape[0]):
    datapt = df.iloc[i]
    start_time = get_sec(datapt['start'])
    end_time = get_sec(datapt['end'])

    link = datapt['link'].strip()
    ids = datapt['link'].split('=')[-1].strip()
    split = 'train'
    if i in test_list:
        split = 'test'
    destination = 'standardized_split_video\\' + split + "\\" + str(i) + datapt['parkinson y/n'].strip()+".mp4"

    
    cmd = "C:\\ffmpeg\\bin\\ffmpeg.exe -ss "+start_time+" -to "+end_time+" -i \"$(youtube-dl --rm-cache-dir -f best --get-url '"+link+"')\" -c:v copy -c:a copy "+destination
    print(cmd)

```
